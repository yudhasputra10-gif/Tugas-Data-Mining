```python
import numpy as np
import pandas as pd
from scipy.spatial.distance import euclidean, cityblock, minkowski
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import jaccard_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.datasets import load_iris
from collections import Counter

# =================================================================
# 1. IMPLEMENTASI JARAK NUMERIK (PROGRAM 1)
# =================================================================
print("=== PROGRAM 1: JARAK NUMERIK ===")
A = np.array([2, 3])
B = np.array([5, 7])

print(f"Euclidean: {euclidean(A, B):.2f}")
print(f"Manhattan: {cityblock(A, B):.2f}")
print(f"Minkowski (p=3): {minkowski(A, B, p=3):.2f}")

# Perbandingan dengan data berskala berbeda
A_scaled = np.array([2, 2000000])
B_scaled = np.array([5, 7000000])
print(f"Tanpa Normalisasi (Euclidean): {euclidean(A_scaled, B_scaled):.2f}")
print("-" * 30)

# =================================================================
# 2. PENTINGNYA NORMALISASI (PROGRAM 2)
# =================================================================
print("\n=== PROGRAM 2: NORMALISASI ===")
data_norm = np.array([[2, 2000000], [5, 7000000], [3, 5000000], [8, 12000000]])

# Min-Max Normalization
scaler_minmax = MinMaxScaler()
data_minmax = scaler_minmax.fit_transform(data_norm)

# Z-Score Normalization
scaler_zscore = StandardScaler()
data_zscore = scaler_zscore.fit_transform(data_norm)

print("Data Setalah Min-Max:\n", data_minmax)
print(f"Jarak Euclidean (Setelah Min-Max): {euclidean(data_minmax[0], data_minmax[1]):.2f}")
print("-" * 30)

# =================================================================
# 3. SIMILARITAS DATA KATEGORIKAL (PROGRAM 3)
# =================================================================
print("\n=== PROGRAM 3: SMC & JACCARD ===")
def simple_matching_coefficient(p, q):
    p, q = np.array(p), np.array(q)
    m11 = np.sum((p == 1) & (q == 1))
    m00 = np.sum((p == 0) & (q == 0))
    return (m11 + m00) / len(p)

p_bin = [1, 0, 1, 1, 0]
q_bin = [1, 0, 0, 1, 1]

print(f"SMC: {simple_matching_coefficient(p_bin, q_bin):.2f}")
print(f"Jaccard (sklearn): {jaccard_score(p_bin, q_bin, average='binary'):.2f}")
print("-" * 30)

# =================================================================
# 4. COSINE SIMILARITY UNTUK TEKS (PROGRAM 4)
# =================================================================
print("\n=== PROGRAM 4: COSINE SIMILARITY TEKS ===")
docs = [
    "data mining machine learning",
    "machine learning data mining",
    "artificial intelligence",
    "deep learning neural network"
]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(docs)
sim_matrix = cosine_similarity(tfidf_matrix)

print("Cosine Similarity Matrix:\n", sim_matrix.round(3))
print("-" * 30)

# =================================================================
# 5. KLASIFIKASI SEDERHANA K-NN (PROGRAM 5)
# =================================================================
print("\n=== PROGRAM 5: K-NN SEDERHANA ===")
def knn_predict(X_train, y_train, X_test, k=3):
    predictions = []
    for test_point in X_test:
        distances = []
        for i, train_point in enumerate(X_train):
            dist = np.sqrt(np.sum((test_point - train_point)**2))
            distances.append((dist, y_train[i]))
        distances.sort(key=lambda x: x[0])
        k_labels = [label for _, label in distances[:k]]
        predictions.append(Counter(k_labels).most_common(1)[0][0])
    return np.array(predictions)

X_train_knn = np.array([[5.1, 3.5], [4.9, 3.0], [6.2, 3.4], [5.9, 3.0]])
y_train_knn = np.array([0, 0, 1, 1])
X_test_knn = np.array([[5.0, 3.2], [6.0, 3.2]])

print(f"Hasil Prediksi k-NN: {knn_predict(X_train_knn, y_train_knn, X_test_knn)}")
print("\n" + "="*50 + "\n")

# =================================================================
# 6. TUGAS RUMAH: IMPLEMENTASI PADA DATASET IRIS
# =================================================================
print("=== TUGAS RUMAH: DATASET IRIS ===")
iris = load_iris()
iris_norm = scaler_minmax.fit_transform(iris.data)

print(f"{'Data':<12} | {'Euclidean':<10} | {'Manhattan':<10} | {'Minkowski':<10}")
print("-" * 55)
for i in range(1, 6):
    e = euclidean(iris_norm[0], iris_norm[i])
    ma = cityblock(iris_norm[0], iris_norm[i])
    mi = minkowski(iris_norm[0], iris_norm[i], p=3)
    print(f"Data 0 - {i} | {e:.4f}     | {ma:.4f}     | {mi:.4f}")

# =================================================================
# 7. SOAL LATIHAN (HITUNGAN MANUAL NOMOR 1, 2, 3)
# =================================================================
print("\n=== JAWABAN SOAL LATIHAN ===")
# No 1
p_lat = np.array([1, 2, 3]); q_lat = np.array([4, 5, 6])
print(f"1. P-Q -> Euclidean: {euclidean(p_lat, q_lat):.4f}, Manhattan: {cityblock(p_lat, q_lat)}")

# No 2
x_lat = [1, 0, 1, 0, 1, 1]; y_lat = [1, 1, 0, 0, 1, 0]
print(f"2. Biner -> SMC: {simple_matching_coefficient(x_lat, y_lat):.4f}, Jaccard: {jaccard_score(x_lat, y_lat):.4f}")

# No 3 (Manual Vector Bag-of-Words)
d1 = [1, 1, 1, 1, 0, 0, 0, 0]; d2 = [1, 0, 1, 1, 1, 0, 0, 0]; d3 = [0, 0, 1, 0, 0, 1, 1, 1]
print(f"3. Cosine -> D1-D2: {cosine_similarity([d1], [d2])[0][0]:.4f}, D1-D3: {cosine_similarity([d1], [d3])[0][0]:.4f}")

```
